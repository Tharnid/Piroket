```python
import pandas as pd
import numpy as np
```

```python
data = np.random.random(size=(5, 3))
print(data)
```

```python
df = pd.DataFrame(data=data, columns=["A", "B", "C"]) # use this method 1
df
```

```python
df = pd.DataFrame(data={"A": [1, 2, 3], "B":["Tharnid", "Saldor", "Aragoth"]}) # use this method 2
df
```

```python
# structured Array
dtype = [("A", np.int), ("B", (np.str, 20))]
data = np.array([(1, "Sam"), (2, "Aragoth"), (3, "Saldor")], dtype=dtype)
print(data)
```
